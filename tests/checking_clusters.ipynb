```python
import pandas as pd

# Assuming 'f' is the h5py File object containing your data
with h5py.File(file_path, 'r') as f:
    # List all keys in the file
    keys = list(f.keys())
    
    # Select keys to exclude 'img_h_latent' and 'img_z_latent'
    exclude_keys = ['img_h_latent', 'img_z_latent']
    selected_keys = [key for key in keys if key not in exclude_keys]
    
    # Extract data for the selected keys and create a DataFrame
    data = {key: f[key][:] for key in selected_keys}
    lung_h5 = pd.DataFrame(data)
    lung_h5.set_index('indexes', inplace=True)
    # Remove byte string 'b' from all columns where necessary
    lung_h5 = lung_h5.applymap(lambda x: x.decode() if isinstance(x, bytes) else x)

# Show the resulting DataFrame
lung_h5
```

```python
merged_df = pd.merge(
    lung_subsample_clinical_captions_clean,
    lung_h5,
    how='inner'
)

merged_df
```

```python
import pandas as pd
import ast
import umap
import matplotlib.pyplot as plt
import seaborn as sns

# Clean img_z_latent column by removing null bytes
merged_df['img_z_latent'] = merged_df['img_z_latent'].apply(lambda x: x.replace(b'\x00', b'') if isinstance(x, bytes) else x)

# Use ast.literal_eval to safely convert the string representation of lists into actual lists
merged_df['img_z_latent'] = merged_df['img_z_latent'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Check if the 'img_z_latent' column is now properly parsed into lists (you can also print some of them for inspection)
print(merged_df['img_z_latent'].head())

# Convert the lists in 'img_z_latent' into a 2D numpy array (a DataFrame where each column represents a feature)
img_z_latent = pd.DataFrame(merged_df['img_z_latent'].to_list())

# Perform UMAP dimensionality reduction
umap_model = umap.UMAP(n_components=2, random_state=42)
umap_result = umap_model.fit_transform(img_z_latent)

# Add UMAP results to the DataFrame
merged_df['umap_1'] = umap_result[:, 0]
merged_df['umap_2'] = umap_result[:, 1]
```

```python
merged_df = merged_df.fillna("Unknown")
merged_df.isna().sum()
```

```python
# Plot UMAP with colors based on Leiden cluster
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='umap_1', y='umap_2', hue='leiden_2.0', data=merged_df, 
    palette='tab20', legend=None, s=10, alpha=0.7
)
plt.title('UMAP of img_z_latent Colored by Leiden Cluster')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the features for the separate plots
features = ['_primary_disease', 'gender', 'tobacco_smoking_history_label']

# Loop over each feature and create a separate UMAP plot
for feature in features:
    plt.figure(figsize=(10, 8))
    
    sns.scatterplot(
        x='umap_1', y='umap_2', hue=feature, data=merged_df,
        palette='Set1', legend='full', s=10, alpha=0.7
    )
    plt.title(f'UMAP of img_z_latent Colored by {feature}')
    
    # Display the plot
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the UMAP plot
plt.figure(figsize=(10, 8))
scatter = sns.scatterplot(
    x='umap_1', y='umap_2', hue='age_at_initial_pathologic_diagnosis', data=merged_df, 
    palette='viridis', legend=None, s=5, alpha=0.7
)

# Set the title for the plot
plt.title('UMAP of img_z_latent Colored by Age')

# Create a colorbar for the 'age' feature
plt.colorbar(scatter.collections[0], label='Age')  # scatter.collections[0] is the mappable object

# Display the plot
plt.show()
```
