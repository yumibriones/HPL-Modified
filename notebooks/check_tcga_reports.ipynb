# Preparing image captions from TCGA Reports

Here we explore the possibility of using TCGA Reports (https://github.com/jkefeli/tcga-path-reports/tree/main) as image captions for CLIP.

```python
import pandas as pd

# import TCGA Reports
lung_clinical = pd.read_csv("/gpfs/home/yb2612/dl4med_25/dl_project/data/scratch_data/TCGA_Reports.csv", header=0)
lung_clinical["sampleID"] = lung_clinical["patient_filename"].str[:12]
lung_clinical.set_index("sampleID", inplace=True)
print("samples, features:", lung_clinical.shape)
lung_clinical
```

```python
import matplotlib.pyplot as plt

# Compute word lengths (split by whitespace)
lung_clinical['word_len'] = lung_clinical['text'].astype(str).apply(lambda x: len(x.split()))

# Display some stats
print("Max word length:", lung_clinical['word_len'].max())
print("Mean word length:", lung_clinical['word_len'].mean())
print("Median word length:", lung_clinical['word_len'].median())

# Plot histogram with vertical line at 60 words
plt.figure(figsize=(10, 6))
plt.hist(lung_clinical['word_len'], bins=50, color='lightgray', edgecolor='black')
plt.axvline(60, color='red', linestyle='dashed', linewidth=2, label='CLIP Limit (~60 words)')
plt.title("Histogram of Report Word Lengths")
plt.xlabel("Word Count")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

## 250k train subset

```python
import h5py
import pandas as pd

file_path = "/gpfs/home/yb2612/dl4med_25/dl_project/data/scratch_data/hdf5_TCGAFFPE_LUADLUSC_5x_60pc_250K_he_train.h5"

with h5py.File(file_path, 'r') as f:
    # Extract and decode string datasets
    tiles = [t.decode('utf-8') for t in f['tiles'][:]]
    slides = [s.decode('utf-8') for s in f['slides'][:]]
    
    # Extract numeric arrays
    labels = f['labels'][:]
    patterns = f['patterns'][:]
    
    # Create DataFrame
    lung_h5 = pd.DataFrame({
        'slides': slides,
        'tiles': tiles,
        'labels': labels,
        'patterns': patterns
    })

    # Add sampleID column (first 15 chars of full_slideID)
    lung_h5['sampleID'] = lung_h5['slides'].str[:12]

    # Set sampleID as index
    lung_h5.set_index('sampleID', inplace=True)

lung_h5
```

```python
# join dfs
lung_images_captions = lung_h5.merge(lung_clinical, left_index=True, right_index=True, how='left')
lung_images_captions
```

```python
import matplotlib.pyplot as plt

# Compute word lengths (split by whitespace)
lung_images_captions['word_len'] = lung_images_captions['text'].astype(str).apply(lambda x: len(x.split()))

# Display some stats
print("Max word length:", lung_images_captions['word_len'].max())
print("Mean word length:", lung_images_captions['word_len'].mean())
print("Median word length:", lung_images_captions['word_len'].median())

# Plot histogram with vertical line at 60 words
plt.figure(figsize=(10, 6))
plt.hist(lung_images_captions['word_len'], bins=50, color='lightgray', edgecolor='black')
plt.axvline(60, color='red', linestyle='dashed', linewidth=2, label='CLIP Limit (~60 words)')
plt.title("Histogram of Report Word Lengths")
plt.xlabel("Word Count")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

There are too many reports with length greater than CLIP's token limit (currently 77). Furthermore, TCGA reports contain information about the pathology of each slide, which would defeat the purpose of HPL's unsupervised clustering approach for identifying HPCs. Thus, we will not proceed with using TCGA reports for CLIP.
