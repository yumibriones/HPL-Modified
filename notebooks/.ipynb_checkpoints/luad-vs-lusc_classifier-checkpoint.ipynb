# Data prep

First we need to save numpy arrays of LUAD labels and Leiden cluster assignments for each of our models and train/test/val splits.

## Extracting LUAD labels

```python
# load h5 file with luad 0/1 labels

import numpy as np
import pandas as pd
import h5py
import os

file_path = "/gpfs/home/yb2612/dl4med_25/dl_project/data/scratch_data/hdf5_TCGAFFPE_LUADLUSC_5x_60pc_he_complete_lungsubtype_survival_filtered.h5"

# Assuming 'f' is the h5py File object containing your data
with h5py.File(file_path, 'r') as f:
    # List all keys in the file
    keys = list(f.keys())
    
    # Select keys to exclude 'img_h_latent' and 'img_z_latent'
    exclude_keys = ['img_h_latent', 'img_z_latent']
    selected_keys = [key for key in keys if key not in exclude_keys]
    
    # Extract data for the selected keys and create a DataFrame
    data = {key: f[key][:] for key in selected_keys}
    lung_h5 = pd.DataFrame(data)
    # lung_h5.set_index('indexes', inplace=True)
    # Remove byte string 'b' from all columns where necessary
    lung_h5 = lung_h5.applymap(lambda x: x.decode() if isinstance(x, bytes) else x)

lung_h5['filepath'] = lung_h5.apply(
        lambda row: f"/gpfs/home/yb2612/dl4med_25/dl_project/data/scratch_data/{row['original_set']}/{row['slides']}/{row['tiles']}",
        axis=1
    )

# Replace 'valid' with 'val' in the 'filepath' column
lung_h5['filepath'] = lung_h5['filepath'].str.replace('valid', 'val')

lung_h5
```

```python
# Create a lookup dictionary for faster matching
filepath_to_luad = {}
for index, row in lung_h5.iterrows():
    filepath_to_luad[row['filepath']] = row['luad']
```

```python
model_names = ["hpl", "clip-scratch", "conch"]
splits = ["test", "train", "val"]

for model_name in model_names:
    for split in splits:
        try:
            print(f"Extracting labels for {model_name}/{split}...")
            filenames = np.load(f"/gpfs/home/yb2612/dl4med_25/dl_project/results/{model_name}/{split}/image_filenames.npy", allow_pickle=True)
            print(f"Loaded {len(filenames)} filenames")
            
            # Extract LUAD labels using the entire filepath
            labels = []
            found_count = 0
            
            for filepath in filenames:
                # Use the entire filepath for matching
                if filepath in filepath_to_luad:
                    # Direct match found
                    label = filepath_to_luad[filepath]
                    labels.append(label)
                    found_count += 1
                else:
                    # Try a more flexible match if direct match fails
                    # This handles potential differences in path formatting
                    matched = False
                    for df_path in filepath_to_luad.keys():
                        if os.path.basename(filepath) == os.path.basename(df_path):
                            label = filepath_to_luad[df_path]
                            labels.append(label)
                            found_count += 1
                            matched = True
                            break
                    
                    if not matched:
                        # Use -1 for not found
                        labels.append(-1)
            
            labels = np.array(labels)
            
            # Count matches found
            print(f"Found labels for {found_count} out of {len(labels)} files")
            
            # Create output directory if it doesn't exist
            os.makedirs(f"/gpfs/home/yb2612/dl4med_25/dl_project/results/{model_name}/{split}", exist_ok=True)
            
            # Save labels
            output_path = f"/gpfs/home/yb2612/dl4med_25/dl_project/results/{model_name}/{split}/image_luad.npy"
            np.save(output_path, labels)
            print(f"Saved labels to {output_path}")
            
        except Exception as e:
            print(f"Error processing {model_name}/{split}: {e}")
```

## Extracting Leiden clusters

```python
import numpy as np
import pandas as pd
import os

# Define models and splits
model_names = ["hpl", "clip-scratch", "conch"]
splits = ["test", "train", "val"]

for model_name in model_names:
    for split in splits:
        print(f"\nProcessing {model_name}/{split}...")
        
        try:
            # Path to image filenames
            filenames_path = f"/gpfs/home/yb2612/dl4med_25/dl_project/results/{model_name}/{split}/image_filenames.npy"
            
            # Path to leiden clustering results
            leiden_path = f"/gpfs/home/yb2612/dl4med_25/dl_project/results/{model_name}/{split}/umap_leiden_results.csv"
            
            # Load image filenames
            filenames = np.load(filenames_path, allow_pickle=True)
            print(f"Loaded {len(filenames)} filenames")
            
            # Load leiden clustering results
            leiden_df = pd.read_csv(leiden_path)
            print(f"Loaded {len(leiden_df)} leiden cluster assignments")
            
            # Create lookup dictionary for faster matching
            filepath_to_leiden = {}
            for index, row in leiden_df.iterrows():
                filepath_to_leiden[row['filepath']] = row['leiden_2.0']
            
            # Extract leiden clusters
            leiden_clusters = []
            found_count = 0
            
            for filepath in filenames:
                # Try direct match first
                if filepath in filepath_to_leiden:
                    cluster = filepath_to_leiden[filepath]
                    leiden_clusters.append(cluster)
                    found_count += 1
                else:
                    # Try matching just the basename
                    matched = False
                    basename = os.path.basename(filepath)
                    
                    for df_path in filepath_to_leiden.keys():
                        if basename == os.path.basename(df_path):
                            cluster = filepath_to_leiden[df_path]
                            leiden_clusters.append(cluster)
                            found_count += 1
                            matched = True
                            break
                    
                    if not matched:
                        # Use -1 for not found
                        leiden_clusters.append(-1)
            
            leiden_clusters = np.array(leiden_clusters)
            
            # Count matches found
            print(f"Found clusters for {found_count} out of {len(leiden_clusters)} files")
            
            # Save leiden clusters
            output_path = f"/gpfs/home/yb2612/dl4med_25/dl_project/results/{model_name}/{split}/image_leiden.npy"
            np.save(output_path, leiden_clusters)
            print(f"Saved leiden clusters to {output_path}")
            
        except FileNotFoundError as e:
            print(f"File not found: {e}")
        except Exception as e:
            print(f"Error: {e}")
```

# Logistic regression classifier

Now we build a classifier that predicts LUAD vs. LUSC based on Leiden cluster assignment.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc

# Simplified load_data function that only loads clusters and labels
def load_data(model_name, data_type):
    """
    Load cluster labels and true labels (LUAD/LUSC) from .npy files
    
    Parameters:
    model_name: str, one of 'hpl', 'clip-scratch', 'conch'
    data_type: str, one of 'train', 'val', 'test'
    
    Returns:
    clusters: numpy array of cluster assignments
    labels: numpy array of true labels (0=LUSC, 1=LUAD)
    """
    base_path = f"/gpfs/home/yb2612/dl4med_25/dl_project/results/{model_name}/{data_type}"
    
    try:
        # Load Leiden cluster assignments
        clusters = np.load(f"{base_path}/image_leiden.npy", allow_pickle=True)
        
        # Load LUAD/LUSC labels
        labels = np.load(f"{base_path}/image_luad.npy", allow_pickle=True)
        
        print(f"Loaded data for {model_name}/{data_type}:")
        print(f"  Clusters shape: {clusters.shape}")
        print(f"  Labels shape: {labels.shape}")
        
        # Filter out any samples with missing labels or clusters
        valid_indices = (labels != -1) & (clusters != -1)
        
        if np.sum(valid_indices) < len(valid_indices):
            print(f"Warning: Filtered {len(valid_indices) - np.sum(valid_indices)} samples with missing labels or clusters")
            clusters = clusters[valid_indices]
            labels = labels[valid_indices]
        
        return clusters, labels
    
    except Exception as e:
        print(f"Error loading data for {model_name}, {data_type}: {e}")
        return None, None

# Function to train and evaluate logistic regression model
def train_evaluate_model(X_train, y_train, X_test, y_test, model_name, feature_type):
    """
    Train a logistic regression model and evaluate its performance
    """
    # Scale features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    # Train logistic regression model
    model = LogisticRegression(max_iter=1000, random_state=42)
    model.fit(X_train_scaled, y_train)
    
    # Evaluate on test set
    y_pred = model.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)
    
    # Get probabilities for AUC calculation
    y_proba = model.predict_proba(X_test_scaled)[:, 1]
    
    print(f"\n--- {model_name} - {feature_type} ---")
    print(f"Accuracy: {accuracy:.4f}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    
    # Create confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['LUSC', 'LUAD'],
                yticklabels=['LUSC', 'LUAD'])
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'{model_name} - {feature_type} Confusion Matrix')
    plt.tight_layout()
    plt.savefig(f'{model_name}_{feature_type}_confusion_matrix.png')
    plt.close()
    
    # Plot ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_name} - {feature_type} ROC Curve')
    plt.legend(loc='lower right')
    plt.savefig(f'{model_name}_{feature_type}_roc_curve.png')
    plt.close()
    
    return model, accuracy, roc_auc

# Function to evaluate how well clusters align with true labels
def evaluate_clusters(clusters, true_labels, model_name):
    """
    Evaluate how well clusters align with true labels
    """
    # Create DataFrame with clusters and true labels
    df = pd.DataFrame({
        'cluster': clusters,
        'label': true_labels
    })
    
    # Calculate statistics for each cluster
    cluster_stats = df.groupby('cluster').agg(
        count=('label', 'count'),
        luad_count=('label', lambda x: sum(x == 1)),
        lusc_count=('label', lambda x: sum(x == 0))
    )
    
    # Calculate percentages
    cluster_stats['luad_percent'] = cluster_stats['luad_count'] / cluster_stats['count'] * 100
    cluster_stats['lusc_percent'] = cluster_stats['lusc_count'] / cluster_stats['count'] * 100
    cluster_stats['dominant_class'] = np.where(
        cluster_stats['luad_percent'] >= 50, 'LUAD', 'LUSC'
    )
    cluster_stats['purity'] = np.maximum(
        cluster_stats['luad_percent'], 
        cluster_stats['lusc_percent']
    )
    
    # Plot cluster purity
    plt.figure(figsize=(12, 6))
    sns.barplot(x=cluster_stats.index, y='purity', hue='dominant_class', data=cluster_stats.reset_index())
    plt.xlabel('Cluster')
    plt.ylabel('Purity (%)')
    plt.title(f'{model_name} - Cluster Purity')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.savefig(f'{model_name}_cluster_purity.png')
    plt.close()
    
    # Print cluster statistics
    print(f"\n--- {model_name} - Cluster Statistics ---")
    print(cluster_stats)
    
    # Calculate cluster-based classification accuracy
    # (Assign each cluster to its majority class and evaluate)
    cluster_to_pred_label = {
        cluster: 1 if row['luad_percent'] > 50 else 0
        for cluster, row in cluster_stats.iterrows()
    }
    
    # Predict labels based on cluster assignments
    cluster_based_preds = np.array([cluster_to_pred_label.get(c, 0) for c in clusters])
    cluster_accuracy = accuracy_score(true_labels, cluster_based_preds)
    
    print(f"\nCluster-based classification accuracy: {cluster_accuracy:.4f}")
    print(classification_report(true_labels, cluster_based_preds))
    
    return cluster_stats, cluster_accuracy
```

```python
# Main execution function
def main():
    # Define models
    models = ['hpl', 'clip-scratch', 'conch']
    
    # Dictionary to store results for comparison
    results = {
        'model': [],
        'feature_type': [],
        'accuracy': [],
        'auc': [],
        'cluster_accuracy': []
    }
    
    for model_name in models:
        print(f"\n\n{'='*50}")
        print(f"Processing {model_name}")
        print(f"{'='*50}")
        
        # Load data for training and testing
        train_clusters, train_labels = load_data(model_name, 'train')
        test_clusters, test_labels = load_data(model_name, 'test')
        
        if train_clusters is None or test_clusters is None:
            print(f"Skipping {model_name} due to data loading error")
            continue
        
        # One-hot encode clusters for model training
        train_clusters_onehot = pd.get_dummies(train_clusters).values
        test_clusters_df = pd.get_dummies(test_clusters)
        
        # Ensure test has same columns as train
        train_clusters_df = pd.get_dummies(train_clusters)
        for col in train_clusters_df.columns:
            if col not in test_clusters_df.columns:
                test_clusters_df[col] = 0
        
        # Keep only the columns from train
        test_clusters_onehot = test_clusters_df[train_clusters_df.columns].values
        
        # Train and evaluate model using clusters
        _, accuracy, auc_score = train_evaluate_model(
            train_clusters_onehot, train_labels,
            test_clusters_onehot, test_labels,
            model_name, 'Clusters Only'
        )
        
        # Evaluate how well clusters align with true labels
        _, cluster_accuracy = evaluate_clusters(
            test_clusters, test_labels, model_name
        )
        
        # Store results
        results['model'].append(model_name)
        results['feature_type'].append('clusters_only')
        results['accuracy'].append(accuracy)
        results['auc'].append(auc_score)
        results['cluster_accuracy'].append(cluster_accuracy)
    
    # Create comparison DataFrame
    results_df = pd.DataFrame(results)
    
    print("\n\n=== Model Comparison ===")
    print(results_df)
    
    # Plot comparison
    plt.figure(figsize=(12, 6))
    
    # Plot accuracy and AUC side by side
    results_melted = pd.melt(results_df, 
                             id_vars=['model', 'feature_type'], 
                             value_vars=['accuracy', 'auc', 'cluster_accuracy'],
                             var_name='metric', value_name='value')
    
    sns.barplot(x='model', y='value', hue='metric', data=results_melted)
    plt.title('Model Performance Comparison')
    plt.ylim(0, 1)
    plt.tight_layout()
    plt.savefig('model_performance_comparison.png')
    
    return results_df

if __name__ == "__main__":
    main()
```
