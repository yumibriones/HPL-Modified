# Accessing images

## One image

```python
import h5py
import matplotlib.pyplot as plt
import numpy as np

file_path = "/gpfs/home/yb2612/dl4med_25/dl_project/data/scratch_data/hdf5_TCGAFFPE_LUADLUSC_5x_60pc_250K_he_train.h5"

with h5py.File(file_path, 'r') as f:
    # Check what's in the 'img' dataset
    img_dataset = f['img']
    print(f"Image dataset shape: {img_dataset.shape}")
    print(f"Image dataset type: {img_dataset.dtype}")
    
    # get the first image (index 0)
    image = img_dataset[0]
    
    # get tile name and slide ID for reference
    tile_name = f['tiles'][0].decode('utf-8')
    slide_id = f['slides'][0].decode('utf-8')
    
    print(f"Displaying image for tile: {tile_name} from slide: {slide_id}")
    
    plt.figure(figsize=(4, 4))
    
    # RGB images
    if len(image.shape) == 3 and image.shape[2] == 3:
        plt.imshow(image)
    # grayscale images
    elif len(image.shape) == 2:
        plt.imshow(image, cmap='gray')
    # different channel structure
    else:
        plt.imshow(np.array(image).astype(np.float32) / np.max(image))
        
    plt.title(f"Tile: {tile_name} | Slide: {slide_id}")
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```

## Multiple images

```python
import h5py
import matplotlib.pyplot as plt
import numpy as np

file_path = "/gpfs/home/yb2612/dl4med_25/dl_project/data/scratch_data/hdf5_TCGAFFPE_LUADLUSC_5x_60pc_250K_he_train.h5"

with h5py.File(file_path, 'r') as f:
    plt.figure(figsize=(4, 8))
    
    # 4 images in a 2x2 grid
    for i in range(4):
        plt.subplot(2, 2, i+1)
        
        image = f['img'][i]
        tile_name = f['tiles'][i].decode('utf-8')
        slide_id = f['slides'][i].decode('utf-8')
        
        # RGB images
        if len(image.shape) == 3 and image.shape[2] == 3:
            plt.imshow(image)
        # grayscale images
        elif len(image.shape) == 2:
            plt.imshow(image, cmap='gray')
        # different channel structure
        else:
            plt.imshow(np.array(image).astype(np.float32) / np.max(image))
            
        plt.title(f"Tile: {tile_name}", fontsize=10)
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```

## Whole slide

```python
import h5py
import numpy as np
from collections import Counter
import pandas as pd

file_path = "/gpfs/home/yb2612/dl4med_25/dl_project/data/scratch_data/hdf5_TCGAFFPE_LUADLUSC_5x_60pc_250K_he_train.h5"

with h5py.File(file_path, 'r') as f:
    # get all slide IDs
    slides = [slide_id.decode('utf-8') for slide_id in f['slides'][:]]
    
    # count occurrences of each
    slide_counts = Counter(slides)
    
    # make df
    slide_df = pd.DataFrame({
        'slide_id': list(slide_counts.keys()),
        'tile_count': list(slide_counts.values())
    })
    
    # sort by tile count
    slide_df = slide_df.sort_values('tile_count')
    
    # find slide with fewest tiles
    min_slide = slide_df.iloc[0]
    
    print(f"Slide with minimum tiles: {min_slide['slide_id']}")
    print(f"Number of tiles: {min_slide['tile_count']}")
    
    print("\nSlides with fewest tiles:")
    print(slide_df.head())
    
    # stats
    print("\nTile count statistics:")
    print(slide_df['tile_count'].describe())
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import os

file_path = "/gpfs/home/yb2612/dl4med_25/dl_project/data/scratch_data/hdf5_TCGAFFPE_LUADLUSC_5x_60pc_250K_he_train.h5"

# slide of interest
target_slide_id = "TCGA-55-8207-01Z-00-DX1"

with h5py.File(file_path, 'r') as f:
    # get slide IDs
    slides = [slide_id.decode('utf-8') for slide_id in f['slides'][:]]
    tiles = [tile_name.decode('utf-8') for tile_name in f['tiles'][:]]
    
    # find indices where slide = slide of interest
    target_indices = [i for i, slide in enumerate(slides) if slide == target_slide_id]
    
    print(f"Found {len(target_indices)} tiles for slide {target_slide_id}")
    
    # get tile names and images
    target_tile_names = [tiles[i] for i in target_indices]
    target_images = [f['img'][i] for i in target_indices]
    
    # display figure
    fig = plt.figure(figsize=(10, 3))
    
    for i, (img, tile_name) in enumerate(zip(target_images, target_tile_names)):
        plt.subplot(1, 3, i+1)
        plt.imshow(img)
        plt.title(f"Tile: {tile_name}")
        plt.axis('off')
    
    plt.suptitle(f"Tiles from slide: {target_slide_id}", fontsize=16)
    plt.tight_layout()
    
    plt.show()
```
